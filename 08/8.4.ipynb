> a) Translate the program into three-address statements of the type we have been using int this section. Assume integers require 4 bytes.

```
 1) i = 2
 2) if i > n goto (7)
 3)     t1 = i * 4
 4)     a[t1] = TRUE
 5)     i = i + 1
 6)     goto (2)
 7) count = 0
 8) s = sqrt(n)
 9) i = 2
10) if i > s goto (22)
11)     t2 = i * 4
12)     ifFalse a[t2] goto (20)
13)         count = count + 1
14)         j = 2 * i
15)         if j > n goto (20)
16)             t3 = j * 4
17)             a[t3] = FALSE
18)             j = j + i
19)             goto (15)
20)     i = i + 1
21)     goto (10)
22)
```
